```python
#2.Mean or Median Imputation Method
from sklearn.datasets import fetch_california_housing
from sklearn.impute import SimpleImputer
```

```python
data=fetch_california_housing()
```

```python
df=pd.DataFrame(data.data,columns=data.feature_names)
```

```python
df.iloc[::10]=np.nan
```

```python
df.isnull().sum()
```

```python
imp=SimpleImputer(strategy='mean')
```

```python
df_imputed=imp.fit_transform(df)
```

```python
df_imputed
```

```python
df_imputed=pd.DataFrame(df_imputed,columns=df.columns)
```

```python
df=df.fillna(df.mean())
```

```python
df
```

```python
from sklearn.datasets import fetch_california_housing
from sklearn.linear_model import LinearRegression
import pandas as pd
import numpy as np
```

```python
data=fetch_california_housing()
df=pd.DataFrame(data.data,columns=data.feature_names)
```

```python
df.head()
```

```python
df.iloc[10:20,0]=np.nan
```

```python
df.head(21)
```

x_missing=df[df.isna().any(axis=1)]
x_no_missing=df.dropna()

```python
x_missing=df[df.isna().any(axis=1)] 
x_no_missing=df.dropna()
```

```python
x_no_missing
```

```python
x_train=x_no_missing.drop(columns=['MedInc'])
```

```python
y_train=x_no_missing['MedInc']
x_test=x_missing.drop(columns=['MedInc'])
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
x_missing['MedInc']=model.predict(x_test)
```

```python
df_final = pd.concat([x_missing,x_no_missing],axis=0)
```
