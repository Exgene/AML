```python
import requests
from bs4 import BeautifulSoup
```

```python
def crawl_web(url):
    res = requests.get(url)

    if res.status_code == 200:
        soup = BeautifulSoup(res.content,'html.parser')

        #Extract relevant data from the page
        #For example let's extract all the links on the page
        links = soup.find_all('a')

        #Print the extracted Links
        for link in links:
            href=link.get('href')
            print(href)

    else:
        print("Request Error: Illegal Link")

```

```python
url = 'https://google.com/'
crawl_web(url)
```

# Write a simple web crawling program to scarp the different paragraphs present in the wikipedia.com/OpenAI

```python
import requests
from bs4 import BeautifulSoup
```

```python
def crawl_web(url):
    res = requests.get(url)

    if res.status_code == 200:
        soup = BeautifulSoup(res.content,'html.parser')

        #Extract relevant data from the page
        #For example let's extract all the links on the page
        para = soup.find_all('p')

        #Print the extracted Links
        for p in para:
            print('--------<->-------')
            print(p.text)

    else:
        print("Request Error: Illegal Link")

```

```python
url = 'https://en.wikipedia.org/wiki/OpenAI'
crawl_web(url)
```
