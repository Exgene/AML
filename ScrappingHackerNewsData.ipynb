```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from datetime import datetime, timedelta
```

```python
url = "https://news.google.com/rss?h1=en-US&gl=US&ceid=US:en&hl=en-US:en"
#Define the URL for GOoogle NEws FEEED 
```

```python
res = requests.get(url)
#Makes a request from the RSS feed URL
```

```python
bs = BeautifulSoup(res.content,'xml')
```

```python
articles = bs.find_all('item')
```

```python
for i in articles:
    print(f'{i.text}\n\n\n')
```

```python
data = []
cutoff_time = datetime.now() - timedelta(hours=24)
```

```python
for article in articles:
    #extract relevant data from each article
    pub_date = datetime.strptime(article.pubDate.text,"%a, %d %b %Y %H:%M:%S %Z")
    #Check if the article was published within the last 24 hours
    if pub_date>=cutoff_time:
        title = article.title.text
        link = article.link.text
        description = article.description.text
        #Append the extracted data to the list
        data.append([title,link,description])

df = pd.DataFrame(data,columns=['Title','Link','Description'])
```

```python
df
```

# Apply unsupervised clustering algorithm to get the labelled data

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
```

```python
#Convert description text into numerical feature
vec = TfidfVectorizer()
x = vec.fit_transform(df['Description'])
```

```python
k = 5
kmeans = KMeans(n_clusters = k, random_state=42)
kmeans.fit(x)
```

```python
labels = kmeans.labels_
```

```python
labels
```

```python
df['Cluster'] = labels
```

```python
#Count the number of articles in each cluster
cluster_counts = df['Cluster'].value_counts().sort_index()
```
