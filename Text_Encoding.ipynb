# Text Encoding
1.Label Encoding:

```python
#Iris Dataset
from sklearn.datasets import load_iris
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
iris = load_iris()
```

```python
iris.data = pd.DataFrame(iris.data, columns = iris.feature_names)
```

```python
iris.data
```

```python
iris.target
```

```python
y = pd.Categorical.from_codes(iris.target, iris.target_names)#Converting the encoded values to Targeted Values
```

```python
y
```

```python
le = LabelEncoder()
```

```python
y = le.fit_transform(y)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(iris.data,y,test_size = 0.7)
```

```python
from sklearn.tree import DecisionTreeClassifier
dcf = DecisionTreeClassifier()
```

```python
dcf.fit(x_train,y_train)
```

```python
dcf.score(x_train,y_train)
```

```python
dcf.score(x_test,y_test)
```

2.One Hot Encoder

```python
iris = load_iris()
```

```python
iris.data = pd.DataFrame(iris.data,columns = iris.feature_names)
```

```python
iris.target
```

```python
y = pd.Categorical.from_codes(y,iris.target_names)
```

```python
y
```

```python
from sklearn.preprocessing import OneHotEncoder
on = OneHotEncoder()
```

```python
y_enc = on.fit_transform(y.reshape(-1,1))
```

```python
y_enc = y_enc.toarray()#Convert from sparse matrix to array
```

```python
x_train,x_test,y_train,y_test = train_test_split(iris.data , y_enc, test_size = 0.25)
```

```python
dcf = DecisionTreeClassifier()
dcf.fit(x_train,y_train)
```

```python
dcf.score(x_test,y_test)
```

3.Word Embedding


```python
import gensim
from gensim.models import Word2Vec
import nltk
nltk.download('punkt')
```

# Steps for Word Embedding
## 1.Tokenize to get individual words from sentence
## 2.LoweCase()
## 3. Removing the Stopwards (Uneccesary words)
## 4.Pre processed words are sent to the Word2Vec
### Vector_Size = How many dimension of a vector for each word
### window = How many previous words to refer
### min_count = frequency
### workers = threads

```python
etext = "text for preprocessing and word embedding using Python and Gensim library"
words = nltk.word_tokenize(etext)
```

```python
processed_words = [word.lower() for word in words if word.isalpha() and word.lower() not in nltk.corpus.stopwords.words()]
```

```python
nltk.download('stopwords')
processed_words
```

```python
model = Word2Vec([processed_words],vector_size=100,window = 5,min_count = 1,workers=8)
```

```python
word = "python"
embed = model.wv[word]
```
