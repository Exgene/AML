# Program 1 to extract title from web page

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://google.com/"
url = "https://github.com/"
response = requests.get(url)
google = BeautifulSoup(response.text,'html.parser')

title = google.find('title').text
print(title)
```

```python
# Extract from nitte website 
url = "https://nmamit.nitte.edu.in/"
response = requests.get(url)
nitte = BeautifulSoup(response.text,'html.parser')

title = nitte.find('title').text
print(title)
```

# Extracting links from the webpage


```python
url1 = "https://nmamit.nitte.edu.in/"
url2 = "https://google.com/"
res1 = requests.get(url1)
res2 = requests.get(url2)

nitte = BeautifulSoup(res1.text,"html.parser")
google = BeautifulSoup(res2.text,"html.parser")
```

```python
for link in nitte.find_all("a"):
    print(link.get("href"))
```

```python
for link in google.find_all("a"):
    print(link.get("href"))
```
