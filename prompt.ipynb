```python
# Text Classifciation models
from transformers import pipeline

# Load the pipeline for text classification
classifier = pipeline('sentiment-analysis')

# Example text
texts = ["I love this product!", "This is the worst experience I've ever had."]

# Classify the text
results = classifier(texts)

# Print results
for text, result in zip(texts, results):
    print(f"Text: {text}\nClassification: {result}\n")
```

```python
# Named Entity recognition
from transformers import pipeline

# Load the pipeline for NER
ner = pipeline('ner', grouped_entities=True)

# Example text
text = "Hugging Face Inc. is a company based in New York City. Its headquarters are in DUMBO, therefore very close to the Manhattan Bridge."

# Recognize named entities
results = ner(text)

# Print results
for entity in results:
    print(f"Entity: {entity['word']}, Type: {entity['entity_group']}, Score: {entity['score']}")
```

```python
# Question Answering
from transformers import pipeline

# Load the pipeline for question answering
qa = pipeline('question-answering')

# Example context and question
context = "Hugging Face Inc. is a company based in New York City. Its headquarters are in DUMBO, therefore very close to the Manhattan Bridge."
question = "Where is Hugging Face's headquarters?"

# Get the answer
result = qa(question=question, context=context)

# Print result
print(f"Question: {question}\nAnswer: {result['answer']}")
```

```python
# Text Generation
from transformers import pipeline

# Load the pipeline for fill-mask
fill_mask = pipeline('fill-mask', model='bert-base-uncased')

# Example text with a masked word
text = "Hugging Face is creating a tool that democratizes AI."

# Predict the masked word
results = fill_mask(text.replace("democratizes", "[MASK]"))

# Print results
for result in results:
    print(f"Prediction: {result['sequence']}, Score: {result['score']}")
```

```python
# Text Summarization
from transformers import pipeline

# Load the pipeline for summarization
summarizer = pipeline('summarization')

# Example long text
text = """
Hugging Face Inc. is a company based in New York City. Its headquarters are in DUMBO, therefore very close to the Manhattan Bridge.
Hugging Face has been creating a number of open-source projects, including the popular 'transformers' library, which has become the go-to tool for natural language processing tasks.
Their mission is to democratize AI by making it accessible to everyone.
"""

# Summarize the text
summary = summarizer(text, max_length=50, min_length=25, do_sample=False)

# Print the summary
print(summary[0]['summary_text'])
```

```python
# Text Translation
from transformers import pipeline

# Load the pipeline for translation
translator = pipeline('translation_en_to_fr')

# Example text
text = "Hugging Face is creating tools that democratize AI."

# Translate the text
translation = translator(text)

# Print the translation
print(translation[0]['translation_text'])
```

```python
# Text Similarity
from transformers import AutoTokenizer, AutoModel
import torch

# Load the tokenizer and model
tokenizer = AutoTokenizer.from_pretrained('bert-base-uncased')
model = AutoModel.from_pretrained('bert-base-uncased')

# Example texts
text1 = "Hugging Face is creating tools that democratize AI."
text2 = "Hugging Face develops open-source projects for natural language processing."

# Tokenize the texts
inputs1 = tokenizer(text1, return_tensors='pt')
inputs2 = tokenizer(text2, return_tensors='pt')

# Get the embeddings
with torch.no_grad():
    embeddings1 = model(**inputs1).last_hidden_state.mean(dim=1)
    embeddings2 = model(**inputs2).last_hidden_state.mean(dim=1)

# Compute cosine similarity
similarity = torch.nn.functional.cosine_similarity(embeddings1, embeddings2)

# Print the similarity
print(f"Similarity: {similarity.item()}")
```
